```python
import random
import numpy
from scipy import stats
```

```python
a=int(input("enter a number"))
j=0
for i in range(1,a+1):
    if a%i==0:
        j=j+1
if j>2:
    print("number is not prime")
else:
    print("number is prime")
```

```python
num=input("enter a number")
str_num = str(num)
if str_num == str_num[::-1]:
    print(f"{num} is a palindrome.")
else:
    print(f"{num} is not a palindrome.")
```

```python
percentage = float(input("Enter your percentage: "))

if percentage >= 90:
    grade = 'A'
elif percentage >= 80:
    grade = 'B'
elif percentage >= 70:
    grade = 'C'
elif percentage >= 60:
    grade = 'D'
elif percentage >= 50:
    grade = 'E'
else:
    grade = 'F'

print(f"Your grade is: {grade}")
```

```python
print("Simple Calculator")
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = input("Enter choice (1/2/3/4): ")


num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == '1':
    result = num1 + num2
    print(f"The result of {num1} + {num2} is: {result}")
elif choice == '2':
    result = num1 - num2
    print(f"The result of {num1} - {num2} is: {result}")
elif choice == '3':
    result = num1 * num2
    print(f"The result of {num1} * {num2} is: {result}")
elif choice == '4':
    if num2 != 0:
        result = num1 / num2
        print(f"The result of {num1} / {num2} is: {result}")
    else:
        print("Error! Division by zero is not allowed.")
else:
    print("Invalid input! Please select a valid operation.")
```

```python
ll=[]
for i in range(25):
    ll.append(random.randint(0,25))
meanll=numpy.mean(ll)
medianll=numpy.median(ll)
modell=stats.mode(ll)
print(ll)
print("mean is", meanll)
print("median is",medianll)
print("mode is",modell)
```

```python
listA=[]
for i in range(1000):
    listA.append(random.randint(0,1000))
print(listA)
listB=[]
for j in range(200):
    listB=[listA[i:i+5] for i in range(0,len(listA),5)]

evenlist=[listB[x] for x in range(0,len(listB)) if x%2==0]
print("even index list",evenlist)
oddlist=[listB[x] for x in range(0,len(listB)) if x%2!=0]
print("odd index list",oddlist)
```
